```python
table = soup.find_all('table')[1]
rows = table.find_all('tr')
row_list = list()
```

```python
for tr in rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    row_list.append(row)
```

```python
print(row_list)
```

```python
df_bs = pd.DataFrame(row_list, columns=['Feature','Value'])
```

```python
print(df_bs)
```
